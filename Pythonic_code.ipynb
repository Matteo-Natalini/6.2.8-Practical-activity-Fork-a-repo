```python
# Import the libraries.
import pandas as pd
import numpy as np

# Import the CSV file.
products = pd.read_csv('Data/products.csv')

# View the DataFrame.
products
```

## # How many products have the word 'Fire' in their description?

```python
# Lambda funtion to find if the word 'fire' is in the DataFrame.
fire_list = products[products['Description'].apply(lambda x: 
                                     'fire' in x.lower())]
# Count of products.
cnt_fire = fire_list['Description'].count() 

# View results.
print("Products with the word 'fire': ", cnt_fire) 

fire_list
```
